```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# Categorical Data

# a. Countplot

```python
sns.countplot(x='Survived', data=df)
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x='Pclass', data=df)
df['Pclass'].value_counts()
```

```python
sns.countplot(x='Sex', data=df)
df['Pclass'].value_counts()
```

```python
sns.countplot(x='Embarked', data=df)
df['Pclass'].value_counts()
```

# b. PieChart

```python
df['Embarked'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2. Numerical Data

# a. Histogram

```python
plt.hist(df['Age'])
```

```python
plt.hist(df['Fare'],bins=10)
```

# b. Distplot 

```python
sns.distplot(df['Age'])
```

```python
# KDE- Kernel Distribution 
sns.kdeplot(df['Age'])
```

```python
# As Distplot is deprecated using Histogram and KDE plot 
plt.figure(figsize=(8, 5))  # Set figure size

# Histogram (like `distplot`) with density normalization
plt.hist(df['Age'], bins=30, density=True, alpha=0.6)

# KDE plot (like `distplot`)
sns.kdeplot(df['Age'])



plt.show()
```

# c. Boxplot

```python
sns.boxplot(df['Fare'])
```

```python
sns.boxplot(df['Age'])
```
